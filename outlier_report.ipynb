```python
# Imports 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
cleaned_central_london = pd.read_csv('cleaned_central_london.csv')
cleaned_inner_london = pd.read_csv('cleaned_inner_london.csv')
cleaned_outer_london = pd.read_csv('cleaned_outer_london.csv')
cleaned_biking_sites = pd.read_csv('cleaned_bike_sites.csv')
```

# Outliers report 

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20, 5))

# Scatterplot for Total cycles
sns.scatterplot(data=cleaned_outer_london, x='Start time', y='Total cycles', ax=ax1)
ax1.set_title('Total cycles')

# Scatterplot for Number of unknown cycles
sns.scatterplot(data=cleaned_outer_london, x='Start time', y='Number of unknown cycles', ax=ax2)
ax2.set_title('Number of unknown cycles')

# Scatterplot for Number of female cycles
sns.scatterplot(data=cleaned_outer_london, x='Start time', y='Number of female cycles', ax=ax3)
ax3.set_title('Number of female cycles')

# Scatterplot for Number of male cycles
sns.scatterplot(data=cleaned_outer_london, x='Start time', y='Number of male cycles', ax=ax4)
ax4.set_title('Number of male cycles')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))

# Scatterplot for Number of private cycles
sns.scatterplot(data=cleaned_inner_london, x='Start time', y='Number of private cycles', ax=ax1)
ax1.set_title('Number of private cycles')

# Scatterplot for Number of cycle hire bikes
sns.scatterplot(data=cleaned_inner_london, x='Start time', y='Number of cycle hire bikes', ax=ax2)
ax2.set_title('Number of cycle hire bikes')

# Scatterplot for Total cycles
sns.scatterplot(data=cleaned_inner_london, x='Start time', y='Total cycles', ax=ax3)
ax3.set_title('Total cycles')

plt.tight_layout()
plt.show()
```
