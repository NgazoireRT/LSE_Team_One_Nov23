 ## NY Bicycle_Routes

```python
biking_routes.head()
```

## Sense Check

```python
missing_values_biking_routes = biking_routes.isnull().sum()
missing_values_biking_routes
```

```python
#Check the missing values
missing_ft_facilit = biking_routes[biking_routes['ft_facilit'].isnull()]
missing_ft_facilit
```

```python
#Check if the NAN(Columns) consistent: against the whole data frame

#Geth the unique: values of the from street and to street
unique_fromstreet_values = missing_ft_facilit['fromstreet'].unique()
unique_tostreet_values = missing_ft_facilit['tostreet'].unique()


# Zip the values: itteriate using for loop
for fromstreet, tostreet in zip(unique_fromstreet_values, unique_tostreet_values):
    #check aginst the whole dataframe using str contains: unique values
    ny_check_miss_data = biking_routes[biking_routes['fromstreet'].str.contains(fromstreet) & (biking_routes['tostreet'] == tostreet)]
    ny_check_miss_data = ny_check_miss_data[~ny_check_miss_data['ft_facilit'].isna()]
    total_sum = ny_check_miss_data.sum()
    print(f'{fromstreet}: {tostreet}, Total Sum: {total_sum}')

```

```python
cleaned_biking_routes= biking_routes.copy()
cleaned_biking_routes
```

```python

# Rename columns
cleaned_biking_routes = cleaned_biking_routes.rename(columns={
    "segmentid": "Segment ID",
    "boro": "Borough",
    "street": "Street Name",
    "fromstreet": "From Street",
    "tostreet": "To Street",
    "bikedir": "Bike Direction",
    "lanecount": "Number of Bike Lanes",
    "ft_facilit": "Bike Facilities"
})
```

```python
cleaned_biking_routes
```

```python
#Fill the na values with unknown.
cleaned_biking_routes = cleaned_biking_routes.fillna("Unknown") 
```
